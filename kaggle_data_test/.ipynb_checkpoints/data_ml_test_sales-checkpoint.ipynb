```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os


import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')
```

```python
my_file = os.getcwd() +  '/data/train.csv'

df = pd.read_csv(my_file)
df = df.set_index('date')
df.index = pd.to_datetime(df.index)

df =df.drop(columns='id')
df
```

```python
store_1 = df.loc[df['store_nbr'] == 1]
store_1 = store_1.drop(columns='store_nbr')
store_1.head()
```

```python
store_1['family'].unique()
```

```python

store_1_cleaning = store_1.loc[ (store_1['family'] == 'CLEANING')]
store_1_cleaning = store_1_cleaning.drop(columns='family')
fig, ax = plt.subplots(figsize=(15, 5))

store_1_cleaning['sales'].plot(ax=ax, label='sales', title='train data')
store_1_cleaning['onpromotion'].plot(ax=ax, label='promo')
plt.show()
```

```python
store_1_cleaning
```

```python
store_1_cleaning.loc[(store_1_cleaning.index > '2015-01-01') & (store_1_cleaning.index < '2015-03-01')] \
    .plot(figsize=(15, 5), title='Week Of Data')
plt.show()
```

```python
def define_topic(df, store_number, topic):
    df = df.copy()
    df = df.loc[df['store_nbr'] == store_number]
    df = df.drop(columns='store_nbr')
    df = df.loc[ (df['family'] == topic)]
    df = df.drop(columns='family')
    return df

store_1_CELEBRATION = define_topic(df, 1, 'CELEBRATION' )
store_1_CELEBRATION
```

```python

def create_features(df):
    """
    Create time series features based on time series index.
    """
    df = df.copy()
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    df['dayofmonth'] = df.index.day
    return df

store_1_cleaning = create_features(store_1_cleaning)

store_1_cleaning
```

```python
store_1_cleaning.dtypes
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.boxplot(data=store_1_cleaning, x='dayofweek', y='sales')
ax.set_title('sales by day')
plt.show()

```

```python
train = store_1_cleaning.loc[store_1_cleaning.index < '2017-01-01']
test = store_1_cleaning.loc[store_1_cleaning.index >= '2017-01-01']

fig, ax = plt.subplots(figsize=(15, 5))
train['sales'].plot(ax=ax, label='Training Set', title='Data Train/Test Split')
test['sales'].plot(ax=ax, label='Test Set')
ax.axvline('2017-01-01', color='black', ls='--')
ax.legend(['Training Set', 'Test Set'])
plt.show()
```

```python
FEATURES = ['dayofyear', 'dayofweek', 'month', 'year', 'dayofmonth', 'onpromotion']
TARGET = 'sales'

X_train = train[FEATURES]
y_train = train[TARGET]

X_test = test[FEATURES]
y_test = test[TARGET]

```

```python


reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                       n_estimators=1000,
                       early_stopping_rounds=50,
                       objective='reg:linear',
                       max_depth=3,
                       learning_rate=0.01)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)
```

```python
fi = pd.DataFrame(data=reg.feature_importances_,
             index=reg.feature_names_in_,
             columns=['importance'])
fi.sort_values('importance').plot(kind='barh', title='Feature Importance')
plt.show()
```

```python
test['prediction'] = reg.predict(X_test)
store_1_cleaning= store_1_cleaning.merge(test[['prediction']], how='left', left_index=True, right_index=True)
ax = store_1_cleaning[['sales']].plot(figsize=(15, 5))
store_1_cleaning['prediction'].plot(ax=ax, style='--')
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Dat and Prediction')
plt.show()
```

```python
ax = store_1_cleaning.loc[(store_1_cleaning.index > '04-01-2017') & (store_1_cleaning.index < '07-08-2017')]['sales'] \
    .plot(figsize=(15, 5), title='Week Of Data')
store_1_cleaning.loc[(store_1_cleaning.index > '04-01-2017') & (store_1_cleaning.index < '07-08-2017')]['prediction'] \
    .plot(style='--')
plt.legend(['Truth Data','Prediction'])
plt.show()
```

```python
score = np.sqrt(mean_squared_error(test['sales'], test['prediction']))
print(f'RMSE Score on Test set: {score:0.2f}')
```

```python
test['error'] = np.abs(test[TARGET] - test['prediction'])

test
```

```python
def add_lags(df):
    target_map = df['sales'].to_dict()
    df['lag1'] = (df.index - pd.Timedelta('364 days')).map(target_map)
    df['lag2'] = (df.index - pd.Timedelta('728 days')).map(target_map)
    df['lag3'] = (df.index - pd.Timedelta('1092 days')).map(target_map)
    return df
```

```python
store_1_cleaning = add_lags(store_1_cleaning)

store_1_cleaning
```

```python
store_1_cleaning = create_features(store_1_cleaning)

FEATURES = ['dayofyear', 'dayofweek', 'month', 'year', 'dayofmonth', 'onpromotion',
            'lag1','lag2','lag3']
TARGET = 'sales'

X_all = store_1_cleaning[FEATURES]
y_all = store_1_cleaning[TARGET]

reg = xgb.XGBRegressor(base_score=0.5,
                       booster='gbtree',    
                       n_estimators=500,
                       objective='reg:linear',
                       max_depth=3,
                       learning_rate=0.01)
reg.fit(X_all, y_all,
        eval_set=[(X_all, y_all)],
        verbose=100)
```

```python
my_file = os.getcwd() +  '/data/test.csv'

future_df = pd.read_csv(my_file)
future_df = future_df.set_index('date')
future_df.index = pd.to_datetime(future_df.index)

future_df =future_df.drop(columns='id')
future_df
```

```python
future_df = define_topic(future_df, 1, 'CLEANING' )
future_df
```

```python
future_df['isFuture'] = True
store_1_cleaning['isFuture'] = False
df_and_future = pd.concat([store_1_cleaning, future_df])
df_and_future = create_features(df_and_future)
df_and_future = add_lags(df_and_future)

df_and_future
```

```python
future_w_features = df_and_future.query('isFuture').copy()
future_w_features
```

```python
future_w_features['pred'] = reg.predict(future_w_features[FEATURES])
future_w_features
```
